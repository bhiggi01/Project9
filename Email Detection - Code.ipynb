The project will look at a number of different Feature Extraction Methods:

  - **Basic Tokenization:** Breaking text down into words or "tokens".
  - **Stop Word Removal:** Removing common words should as "the", "and" and "a".
  - **Stemming:** Reduce words to their root forms, such as "run", "running" and "runner" all being reduce to their root form of run. QUESTION: Does this reduce context.
  - **Term Frequency-inverse document frequency (TF-IDF):** assign weights to words based on their frequency in a document.

  <br>
  - **Part of Speech (POS) tagging:** Assigning a grammatical category to each word in  a sentence such as a verb, noun, adjective.
  - **Name entity recongnition (NER):** labelling specfic named entities such as people, locations and organisations.
  - **Word Embeddings:** REpresenting words as high dimensional vectors that capture their meaning and context, often learned from large ammounts of data with Neural Networks.

## **Method 2: Tokenize with Count Vectorization**

Tokenization is the process of breaking down the emails into indiviudal words or tokens. Often the first step in NLP for preparing text for analysis.

These tokens are the basic units that are fed into the NLP model for further processing. Using white space is the most common method to tokenize. Other Methods include punctuation marks, contractions and hypenated words.

```python
# Take raw files and make a version one for tokenising
X_train_token = X_train_raw
y_train_token = y_train_raw
X_val_token = X_val_raw   
y_val_token = y_val_raw
X_test_token = X_test_raw
y_test_token = y_test_raw
```

```python
# create a countVerctorizer
vectorizer_token = CountVectorizer()

X_train_token_vect = vectorizer_token.fit_transform(X_train_token)
X_val_token_vect = vectorizer_token.transform(X_val_token)
X_test_token_vect = vectorizer_token.transform(X_test_token)
```

```python
# shape of the files
X_train_token_vect.shape, X_val_token_vect.shape, X_test_token_vect.shape
```

```python
# Ssze of the sparse matrix
X_train_token_vect, X_val_token_vect, X_test_token_vect
```

## **Method 4: Stemming**

Stemming is the process of reducing words to its root base form, which is a stem. So "running", "ran" and "runner" are reduced to "run. Stemming can be used to improve the efficiency of text analysis by grouping together words of similar meaning which reduces the size of the vocabulary that needs to be processed which then improves the accuracy of text classification. Lets see how it compares to Method 1 and Method 2.

We will also build on previous methods so will contuine to use Stop Words.

### Function to Remove punctuations, word "Subject" and lower case words

I orgionally had planned to do this in an earlier section but I wanted to start Feature Extraction to look at how these items came up so that is the reason for coming up now.
```python
def processing_update(email):
    
    # change all words to lowercase
    email = email.lower()
    
    # remove subject liens
    email = re.sub(r"subject: ", "", email)
    
    # remove punctuation marks
    punctuation = string.punctuation.replace("@", "") # Keep the @ becuase a certain address may be sending spam emails
    translator = str.maketrans("", "", punctuation)
    email = email.translate(translator)
    
    # remove start and end white spaces
    email = email.strip()
        
    return email
```

## **Method 5: Term Frequency-inverse document frequency (TF-IDF)**

TF-IDF reflects the importance of a word by giving it a weight to the words that occure more often. It uses two metrics:

   - Term Frequency (TF): This is the number of times a word appears divided by the total number of words in a document. shows the importance of a word in a certain document.
   - Inverse Document Frequency (IDF): Looks at the total number of documents that contain the word. Shows the rarity of a word in the entire corpus.
    
I could use TF-IDF seperatley but I want to build this on previous methods.

```python
# Take raw files and make a version one for stemming
X_train_tf_idf = X_train_raw2
y_train_tf_idf = y_train_raw2
X_val_tf_idf = X_val_raw2   
y_val_tf_idf = y_val_raw2
X_test_tf_idf = X_test_raw2
y_test_tf_idf = y_test_raw2
```
